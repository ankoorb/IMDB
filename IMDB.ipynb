### TV Shows with 1000+ Votes
##### Link: http://www.imdb.com/search/title?at=0&num_votes=1500,&sort=num_votes&start=1&title_type=tv_series

```python
# Movies with 10000+ Votes: Information Web Scraping
title = []
year = []
text = []
genre = []
runtime = []
rating = []
votes = []
for i in range(1, 6600, 50):
    vote_url = 'http://www.imdb.com/search/title?at=0&num_votes=10000,&sort=num_votes,desc&start='+str(i)
    r_vote = requests.get(vote_url)
    bs = BeautifulSoup(r_vote.text)

    for movie in bs.findAll('td', 'title'):
        # Title
        title.append(movie.find('a').contents[0])
        # Year
        year.append(movie.find('span', 'year_type').contents[0])
        # Rating
        rating.append(movie.find('span', 'value').contents[0])
        # Runtime
        if movie.find('span', 'runtime') is None:
            runtime.append('')
        else:
            runtime.append(movie.find('span', 'runtime').contents[0])
        # Genre 
        if movie.find('span', 'genre') is None:
            genre.append('')
        else:
            genres = movie.find('span', 'genre').findAll('a')
            genre.append([g.contents[0] for g in genres])  
        # Text 
        if movie.find('span', 'outline') is None:
            text.append('')
        else:
            text.append(movie.find('span', 'outline').contents[0])
        
    # Votes
    for movie in bs.findAll('td', 'sort_col'):
        votes.append(movie.contents[0])
```

```python
# Movies with 10000+ votes: Box-Office Revenue Web Scraping
bo_title = []
bo_year = []
bo_runtime = []
bo_rating = []
revenue = []

for i in range(1, 6600, 50):
    url = 'http://www.imdb.com/search/title?at=0&num_votes=10000,&sort=boxoffice_gross_us&start='+str(i)
    r = requests.get(url)
    bs = BeautifulSoup(r.text)

    for movie in bs.findAll('td', 'title'):
        # Title
        bo_title.append(movie.find('a').contents[0])
        # Year
        bo_year.append(movie.find('span', 'year_type').contents[0])
        # Rating
        bo_rating.append(movie.find('span', 'value').contents[0])
        # Runtime
        if movie.find('span', 'runtime') is None:
            bo_runtime.append('')
        else:
            bo_runtime.append(movie.find('span', 'runtime').contents[0])
        
        
    # Votes
    for movie in bs.findAll('td', 'sort_col'):
        revenue.append(movie.contents[0])
```

```python
# TV Shows with 1500+ votes
s_title = []
s_year = []
s_text = []
s_genre = []
s_runtime = []
s_rating = []
s_votes = []
for i in range(1, 2000, 50):
    url = 'http://www.imdb.com/search/title?at=0&num_votes=1500,&sort=num_votes&start='+str(i)+'&title_type=tv_series'
    r = requests.get(url)
    bs = BeautifulSoup(r.text)

    for show in bs.findAll('td', 'title'):
        # Title
        s_title.append(show.find('a').contents[0])
        # Year
        s_year.append(show.find('span', 'year_type').contents[0])
        # Rating
        s_rating.append(show.find('span', 'value').contents[0])
        # Runtime
        if show.find('span', 'runtime') is None:
            s_runtime.append('')
        else:
            s_runtime.append(show.find('span', 'runtime').contents[0])
        # Genre 
        if show.find('span', 'genre') is None:
            s_genre.append('')
        else:
            genres = show.find('span', 'genre').findAll('a')
            s_genre.append([g.contents[0] for g in genres])  
        # Text 
        if show.find('span', 'outline') is None:
            s_text.append('')
        else:
            s_text.append(show.find('span', 'outline').contents[0])
        
    # Votes
    for show in bs.findAll('td', 'sort_col'):
        s_votes.append(show.contents[0])
```

```python
# Movie Vote DataFrame
movie_vote_cols = ['title', 'year', 'runtime', 'rating', 'votes', 'genre', 'text']
movie_votes = pd.DataFrame({'title': title, 'year': year, 'runtime': runtime, 'rating': rating, 'votes': votes, 'genre': genre, 'text': text}, columns = movie_vote_cols)
movie_votes.head(3)
```

```python
# Movie Revenue DataFrame
movie_revenue_cols = ['title', 'year', 'runtime', 'rating', 'revenue']
movie_revenue = pd.DataFrame({'title': bo_title, 'year': bo_year, 'runtime': bo_runtime, 'rating': bo_rating, 'revenue': revenue}, columns = movie_revenue_cols)
movie_revenue.head(3)
```

```python
# TV Show DataFrame
tv_show_cols = ['title', 'year', 'runtime', 'rating', 'votes', 'genre', 'text']
tv_shows = pd.DataFrame({'title': s_title, 'year': s_year, 'runtime': s_runtime, 'rating': s_rating, 'votes': s_votes, 'genre': s_genre, 'text': s_text}, columns = tv_show_cols)
tv_shows.head(3)
```

```python
# Save DataFrames to *.csv
movie_votes.to_csv('movie_votes.csv')
movie_revenue.to_csv('movie_revenue.csv')
tv_shows.to_csv('tv_shows.csv')
```

```python
# Preparing Data For ML
temp = movies[['rating', 'year', 'runtime', 'votes', 'revenue']]
```

```python
# Splitting data: Train = 70% and Test = 30%
mask = np.random.rand(len(temp)) < 0.7
train = temp[mask]
test = temp[~mask]
```

```python
from sklearn.linear_model import LinearRegression
train_X = train.drop('rating', axis = 1)
train_Y = train['rating']
# Create a LinearRegression object
lm = LinearRegression()
lm
```

```python
# Fitting linear regression model
lm.fit(train_X, train_Y)
```

```python
# Coefficients of the model
pd.DataFrame(zip(train_X.columns, lm.coef_), columns = ['Feature', 'Estimated Coefficient'])
```

```python
print 'Estimated intercept coefficient: ', lm.intercept_
```
